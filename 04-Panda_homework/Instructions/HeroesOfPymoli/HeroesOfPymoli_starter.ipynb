### Heroes Of Pymoli Data Analysis
* Of the 1163 active players, the vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).

* Our peak age demographic falls between 20-24 (44.8%) with secondary groups falling between 15-19 (18.60%) and 25-29 (13.4%).

```python
# Dependencies and Setup
import pandas as pd

# Reading the Heroes of Pymoli purchase_data.csv
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)
purchase_data_df.head(20)
#Dataframe to hold only three columns, this will eventually will not have duplicate "SN"
demographics_data_df = purchase_data_df[["SN","Age","Gender"]]

#Debug
demographics_data_df.head()
purchase_data_df.head()
```

## Player Count

* Display the total number of players


```python
#Removing duplicate "SN", and storing in the same data frame
demographics_data_df = demographics_data_df.drop_duplicates(subset="SN")
#obtain the total count of players
total_players = demographics_data_df["Gender"].value_counts().sum()

#Creating a dataframe to display Player Count
player_count_df = pd.DataFrame({"Total Players": [total_players]})
player_count_df


```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#using original purchase_data dataframe which has duplicate "SN"
#obtaining the count of unique Items.
unique_items = len(purchase_data_df["Item ID"].unique())
#Average Price
average_price = purchase_data_df["Price"].mean()
#Total no of Purchases
total_purchases = purchase_data_df["Purchase ID"].count()
#Total Revenue
total_revenue = purchase_data_df["Price"].sum()

#Creating a summary df to display Purchasing Analysis(Total)
purchase_summary_df = pd.DataFrame({"Number of Unique Items": [unique_items],
                                    "Average Price": [average_price],
                                   "Number of Purchases": [total_purchases],
                                   "Total Revenue": [total_revenue]})
#Formating the data for display
purchase_summary_df["Average Price"] = purchase_summary_df["Average Price"].map("${:.2f}".format)
purchase_summary_df["Total Revenue"] = purchase_summary_df["Total Revenue"].map("${:.2f}".format)
purchase_summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Obtaining Gender Demographics without duplicate "SN", else the count will be incorrect
#Grouping by Gender
gender_group = demographics_data_df.groupby("Gender")
#Declaring a variable as its needed in Purchase Analysis(gender) later
non_duplicate_gender_count = gender_group["Gender"].count()
#Creating a DataFrame to display Gender Demographics
gender_demograph_df = pd.DataFrame({"Total Count" : non_duplicate_gender_count})
gender_demograph_df["Percentage"] = (gender_demograph_df.loc
                                     [:, ["Total Count"]]/gender_demograph_df["Total Count"].sum())*100

#Formating data for display
gender_demograph_df["Percentage"] = gender_demograph_df["Percentage"].map("{:.2f}".format)
#Sorting the Dataframe by Total Count of each gender
gender_demograph_df = gender_demograph_df.sort_values("Total Count",ascending=False)

gender_demograph_df.head()

```


## Purchasing Analysis (Gender)

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python

#Creating Age Bins
age_bins = [0,9,14,19,24,29,34,39,49]
#Creating Age Labels
age_labels = ["<10", "10-14", "15-19","20-24","25-29","30-34","35-39","40+"]
 

#Obtaining Gender Demographics without duplicate "SN", else the count will be incorrect
#Making a copy of demographics data and sorting by Age
age_demographics = demographics_data_df.sort_values("Age",ascending=True)
#Using the cut function to group the data in bins and assigning it into a column in the data frame
age_demographics ["Age Groups"] = pd.cut(demographics_data_df["Age"], age_bins, labels=age_labels)

#Now grouping the data by Age Groups
age_groups = age_demographics.groupby("Age Groups")
age_groups.count()

#Creating a summary data frame to display age demographics
#Declaring variable for age_grp_gender_count and will be used now and later for purchase analysis
age_grp_gender_count = age_groups["Gender"].count()
disp_age_demo_df = pd.DataFrame({"Total Count" : age_grp_gender_count })
disp_age_demo_df["Percentage of Players"]= (disp_age_demo_df["Total Count"]/total_players)*100

#Formating data for display
disp_age_demo_df["Percentage of Players"] = disp_age_demo_df["Percentage of Players"].map("{:.2f}".format)

disp_age_demo_df

```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#using original purchase_data and doing a group by "SN"
top_spender_group = purchase_data_df.groupby("SN")

#Display Top Spender data
top_spenders_df = pd.DataFrame({"Purchase Count" : top_spender_group["Purchase ID"].count()})
top_spenders_df["Average Purchase Price"] = top_spender_group["Price"].sum()/top_spender_group["Purchase ID"].count()
top_spenders_df["Total Purchase Value"] = top_spender_group["Price"].sum()

#Sorting data on Total Purchase Value to show top spender
top_spenders_df = top_spenders_df.sort_values("Total Purchase Value",ascending=False)
#Formating data 
top_spenders_df["Average Purchase Price"] = top_spenders_df["Average Purchase Price"].map("${:.2f}".format)
top_spenders_df["Total Purchase Value"] = top_spenders_df["Total Purchase Value"].map("${:.2f}".format)

top_spenders_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Obtaining only three columns from original data frame
popularItems_df = purchase_data_df[["Item ID","Item Name","Price"]]

#Group by on two columns
popularItems_grp = popularItems_df.groupby(["Item ID","Item Name"])

#Declaring variables
purchase_count = popularItems_grp["Item ID"].count()
total_purchase_val = popularItems_grp["Price"].sum()

#Creating a summary dataframe for display of Most Populat items

most_popular_df = pd.DataFrame({"Purchase Count" : purchase_count})
#was not sure if it was okay to use unique and hence used total/count to arrive at individual price
#most_popular_df["Item Price1"] = popularItems_grp["Price"].unique()
most_popular_df["Item Price"] = total_purchase_val/purchase_count
most_popular_df["Total Purchase Value"] = total_purchase_val

#Sorting on Purchase Count to arrive at most popular items
most_popularItems_df = most_popular_df.sort_values("Purchase Count",ascending=False)

#Formating data 
most_popularItems_df["Item Price"] = most_popularItems_df["Item Price"].map("${:.2f}".format)
most_popularItems_df["Total Purchase Value"] = most_popularItems_df["Total Purchase Value"].map("${:.2f}".format)

most_popularItems_df.head()
```
