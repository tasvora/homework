```python
%matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
crime2017_csv = "../Resources/OhioCrime2017_NoNan.csv"
```

```python

crime2017_df = pd.read_csv(crime2017_csv)
crime2017_df = crime2017_df.fillna(0)
crime2017_df["POPULATION"] = pd.to_numeric(crime2017_df["POPULATION"])
crime2017_df = crime2017_df.sort_values("POPULATION",ascending=False)
top20_crime2017_df = crime2017_df.iloc[0:20]


top20_crime2017_df
```

```python
grouped_crime_df1 = top20_crime2017_df.set_index(["City"])

grouped_crime_df = top20_crime2017_df.groupby(['City'])

sumCrime = grouped_crime_df["VIOLENT CRIME"].sum() + grouped_crime_df["MURDER"].sum() + grouped_crime_df["RAPE"].sum() + \
            grouped_crime_df["ROBBERY"].sum() + grouped_crime_df["AGGRAVATED ASSAULT"].sum()

Populate = grouped_crime_df["POPULATION"]

    
```

```python
crime_summary_table = pd.DataFrame({"sumCrime": sumCrime, "population" : Populate})
crime_summary_table.head()
#crime_summary_table["Crime%"] = sumCrime*100/grouped_crime_df["POPULATION"]
#grouped_crime_df.head()
```
